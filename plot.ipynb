```python
import numpy as np
import matplotlib.pyplot as plt

b_a_o = np.array([6.3])
b_a_o_t = np.array([4.2])
b_a_o_t = np.cumsum(b_a_o_t)


b_a_e = np.array([6.3])
b_a_e_t = np.array([2.5])
b_a_e_t = np.cumsum(b_a_e_t)
b_a_e_t_l = np.array([])



b_an_o = np.array([10.2, 6.3])
b_an_o_t = np.array([1.16, 1.8])
b_an_o_t = np.cumsum(b_an_o_t)


b_an_e = np.array([11.1, 7.3, 6.3])
b_an_e_t = np.array([1.7, 1.4, 1.1])
b_an_e_t = np.cumsum(b_an_e_t)

btmax = np.max(np.array([max(b_a_o_t), max(b_a_e_t), max(b_an_e_t), max(b_an_o_t)]))
bmax = np.max(np.array([max(b_a_o), max(b_a_e), max(b_an_e), max(b_an_o)]))


m_a_o = np.array([6])
m_a_o_t = np.array([49.4])
m_a_o_t = np.cumsum(m_a_o_t)

m_a_e = np.array([5.9])
m_a_e_t = np.array([65.8])
m_a_e_t = np.cumsum(m_a_e_t)

m_an_o = np.array([7.1, 6.3])
m_an_o_t = np.array([9.2, 75.2])
m_an_o_t = np.cumsum(m_an_o_t)

m_an_e = np.array([12.9, 6.9, 6.2])
m_an_e_t = np.array([21.9, 18.3, 15.7])
m_an_e_t = np.cumsum(m_an_e_t)

mtmax = np.max(np.array([max(m_a_o_t), max(m_a_e_t), max(m_an_e_t), max(m_an_o_t)]))
mmax = np.max(np.array([max(m_a_o), max(m_a_e), max(m_an_e), max(m_an_o)]))


d_a_o = np.array([10.2])
d_a_o_t = np.array([23.2])
d_a_o_t = np.cumsum(d_a_o_t)

d_a_e = np.array([10.3])
d_a_e_t = np.array([29.5])
d_a_e_t = np.cumsum(d_a_e_t)

d_an_o = np.array([10.4])
d_an_o_t = np.array([3.7])
d_an_o_t = np.cumsum(d_an_o_t)

d_an_e = np.array([10.4])
d_an_e_t = np.array([7.6])
d_an_e_t = np.cumsum(d_an_e_t)

dtmax = np.max(np.array([max(d_a_o_t), max(d_a_e_t), max(d_an_e_t), max(d_an_o_t)]))
dmax = np.max(np.array([max(d_a_o), max(d_a_e), max(d_an_e), max(d_an_o)]))

plt.vlines( b_a_o_t, b_a_o, bmax+5, label='Octile, A*', colors='k')
plt.vlines( b_a_e_t, b_a_e, bmax+5, label='Eucledian, A*', colors='b')
for m, i in enumerate(b_an_o):
    if m==0:
        plt.vlines(b_an_o_t[m], bmax+5, i, colors='r')
    if m==len(b_an_o)-1:
        plt.hlines(i, b_an_o_t[m], btmax, colors='r', label='Octile ANA*')
    if m>=0 and m<len(b_an_o)-1:
        plt.hlines(i, b_an_o_t[m], b_an_o_t[m+1], colors='r')
    if m>0 and m<=len(b_an_o)-1:
        plt.vlines(b_an_o_t[m], i, b_an_o[m-1], colors='r')
for m, i in enumerate(b_an_e):
    if m==0:
        plt.vlines(b_an_e_t[m], bmax+5, i, colors='g')
    if m==len(b_an_e)-1:
        plt.hlines(i, b_an_e_t[m], btmax+5, colors='g', label='Eucledian ANA*')
    if m>=0 and m<len(b_an_e)-1:
        plt.hlines(i, b_an_e_t[m], b_an_e_t[m+1], colors='g')
    if m>0 and m<=len(b_an_e)-1:
        plt.vlines(b_an_e_t[m], i, b_an_e[m-1], colors='g')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Costs')
plt.title("Env: Blocked")
plt.xlim(0, btmax+1)
plt.ylim(0, bmax+2)
plt.show()

plt.vlines( m_a_o_t, m_a_o, mmax+5, label='Octile, A*', colors='k')
plt.vlines( m_a_e_t, m_a_e, mmax+5, label='Eucledian, A*', colors='b')
for m, i in enumerate(m_an_o):
    if m==0:
        plt.vlines(m_an_o_t[m], mmax+5, i, colors='r')
    if m==len(m_an_o)-1:
        plt.hlines(i, m_an_o_t[m], mtmax, colors='r', label='Octile ANA*')
    if m>=0 and m<len(m_an_o)-1:
        plt.hlines(i, m_an_o_t[m], m_an_o_t[m+1], colors='r')
    if m>0 and m<=len(m_an_o)-1:
        plt.vlines(m_an_o_t[m], i, m_an_o[m-1], colors='r')
for m, i in enumerate(m_an_e):
    if m==0:
        plt.vlines(m_an_e_t[m], mmax+5, i, colors='g')
    if m==len(m_an_e)-1:
        plt.hlines(i, m_an_e_t[m], mtmax+5, colors='g', label='Eucledian ANA*')
    if m>=0 and m<len(m_an_e)-1:
        plt.hlines(i, m_an_e_t[m], m_an_e_t[m+1], colors='g')
    if m>0 and m<=len(m_an_e)-1:
        plt.vlines(m_an_e_t[m], i, m_an_e[m-1], colors='g')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Costs')
plt.title("Env: MoveSev")
plt.xlim(0, mtmax+1)
plt.ylim(0, mmax+2)
plt.show()

plt.vlines( d_a_o_t, d_a_o, dmax+5, label='Octile, A*', colors='k')
plt.vlines( d_a_e_t, d_a_e, dmax+5, label='Eucledian, A*', colors='b')
for m, i in enumerate(d_an_o):
    if m==0:
        plt.vlines(d_an_o_t[m], dmax+5, i, colors='r')
    if m==len(d_an_o)-1:
        plt.hlines(i, d_an_o_t[m], dtmax, colors='r', label='Octile ANA*')
    if m>=0 and m<len(d_an_o)-1:
        plt.hlines(i, d_an_o_t[m], d_an_o_t[m+1], colors='r')
    if m>0 and m<=len(d_an_o)-1:
        plt.vlines(d_an_o_t[m], i, d_an_o[m-1], colors='r')
for m, i in enumerate(d_an_e):
    if m==0:
        plt.vlines(d_an_e_t[m], dmax+5, i, colors='g')
    if m==len(d_an_e)-1:
        plt.hlines(i, d_an_e_t[m], dtmax+5, colors='g', label='Eucledian ANA*')
    if m>=0 and m<len(d_an_e)-1:
        plt.hlines(i, d_an_e_t[m], d_an_e_t[m+1], colors='g')
    if m>0 and m<=len(d_an_e)-1:
        plt.vlines(d_an_e_t[m], i, d_an_e[m-1], colors='g')
plt.legend(loc='lower left')
plt.title("Env: PR2Doorway")
plt.xlabel('Time')
plt.ylabel('Costs')
plt.xlim(0, dtmax+1)
plt.ylim(0, dmax+2)
plt.show()
```
